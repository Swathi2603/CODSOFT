```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/Titanic-Dataset.csv")
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
target=df.Survived
input=df.drop('Survived',axis=1)
d=pd.get_dummies(input.Sex)
input=pd.concat([input,d],axis=1)
input.drop('Sex',axis=1,inplace=True)
input.Age=input.Age.fillna(input.Age.mean())
input.head()

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input,target,test_size=0.1)
from sklearn.naive_bayes import GaussianNB
m=GaussianNB()
m.fit(x_train, y_train)
m.score(x_test,y_test)
x_test[:10]
```

```python
a=m.predict(x_test[:20])
for i in a:
  if(i==0):
    print("Not Survived")
  else:
    print("Survived")
```
